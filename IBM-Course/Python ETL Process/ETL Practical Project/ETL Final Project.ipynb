```python
def extract():
    # Write your code here
    extracted_data = extract_from_json('Data/bank_market_cap_1.json')
    # Keep only the specified columns
    extracted_data = extracted_data[['Name', 'Market Cap (US$ Billion)']]
    return extracted_data
```

```python
# Write your code here
exchange_rates = pd.read_csv('Data/exchange_rates.csv', index_col=0)
exchange_rate = exchange_rates.loc['GBP', 'Rates']
print(f"GBP exchange rate: {exchange_rate}")
```

```python
def transform(data, exchange_rate):
    # Write your code here
    # 1. Convert USD to GBP
    data['Market Cap (GBP$ Billion)'] = data['Market Cap (US$ Billion)'] * exchange_rate
    
    # 2. Round to 3 decimal places
    data['Market Cap (GBP$ Billion)'] = data['Market Cap (GBP$ Billion)'].round(3)
    
    # 3. Drop the original USD column
    data = data.drop(columns=['Market Cap (US$ Billion)'])
    
    return data
```

```python
def load(data, output_file):
    # Write your code here
    data.to_csv(output_file, index=False)
```

```python
def log(message):
    # Write your code here
    timestamp_format = '%Y-%m-%d %H:%M:%S'
    now = datetime.now()
    timestamp = now.strftime(timestamp_format)
    with open("logfile.txt", "a") as f:
        f.write(timestamp + ',' + message + '\n')
```

```python
# Call the function here
df_transformed = transform(df_extracted, exchange_rate)

# Print the first 5 rows here
print("\nFirst 5 rows of transformed data:")
print(df_transformed.head())
```
